```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Define the binary cross-entropy loss
def binary_cross_entropy_loss(z, y_true):
    y_pred = sigmoid(z)
    return - (y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

# Generate values for the weighted sum of features (z)
z_values = np.linspace(-10, 10, 500)

# Calculate the loss for true labels y=1 and y=0
loss_y1 = binary_cross_entropy_loss(z_values, y_true=1)
loss_y0 = binary_cross_entropy_loss(z_values, y_true=0)

# Plot the loss as a function of z
plt.figure(figsize=(8, 6))
plt.plot(z_values, loss_y1, label="y=1", linestyle='-', linewidth=2)
plt.plot(z_values, loss_y0, label="y=0", linestyle='--', linewidth=2)
plt.title("Binary Cross-Entropy Loss vs Weighted Sum (z)", fontsize=14)
plt.xlabel("Weighted Sum of Features (z)", fontsize=12)
plt.ylabel("Loss", fontsize=12)
plt.axhline(0, color="gray", linestyle="--", linewidth=0.5)
plt.axvline(0, color="gray", linestyle="--", linewidth=0.5)
plt.legend(fontsize=12)
plt.grid(alpha=0.5)
plt.show()
```
